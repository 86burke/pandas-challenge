```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculations Distric Summary
total_num_schools = school_data_complete['school_name'].nunique()
#print("Total Number of Schools: " + str(total_num_schools))--Testing
total_num_students = school_data['size'].sum()
#print("Total Number of Students: " + str(total_num_students))--Testing
total_budget = school_data['budget'].sum()
#print("Total Budget: " + str(total_budget))--Testing
avg_math = school_data_complete['math_score'].mean()
#print("Average Math Score: " + str(avg_math))--Testing
avg_reading = school_data_complete['reading_score'].mean()
#print("Average Reading Score: " + str(avg_reading))--Testing
passing_math = len(school_data_complete[school_data_complete['math_score'] >=70])
percent_passing_math = passing_math / total_num_students * 100
#print("Percent of Sudents Passing Math: " + str(passing_math))--Testing
passing_reading = len(school_data_complete[school_data_complete['reading_score'] >=70])
percent_passing_reading = passing_reading / total_num_students * 100
#print("Percent of Sudents Passing Reading: " + str(passing_reading))--Testing
overall_percent_passing = (avg_math + avg_reading) /2
#print("Percent Overall Passing Rate: " + str(overall_percent_passing))--Testing
#Create dataframe
dist_summary = pd.DataFrame({"Total Number of Schools": [total_num_schools], "Total Number of Students": [total_num_students]
                            , "Total Budget": [total_budget], "Average Math Score": [avg_math], "Average Reading Score": [avg_reading]
                            ,"Percent of Sudents Passing Math": [percent_passing_math], "Percent of Sudents Passing Reading": [percent_passing_reading]
                            ,"Percent Overall Passing Rate": [overall_percent_passing]})
dist_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#groupby school
school_group = school_data_complete.groupby(["school_name"])
school_type = school_group["type"].first()
total_students = school_group["Student ID"].count()
total_budget2 = school_group["budget"].first()
per_student_budget = total_budget2 / total_students
avg_math2 = school_group["math_score"].mean()
avg_reading2 = school_group["reading_score"].mean()
passing_math2 = school_data_complete[school_data_complete["math_score"] >=70].groupby(["school_name"])["math_score"].count()
percent_passing_math2 = passing_math2 / total_students * 100
passing_reading2 = school_data_complete[school_data_complete["reading_score"] >=70].groupby(["school_name"])["reading_score"].count()
percent_passing_reading2 = passing_reading2 / total_students * 100
overall_percent_passing2 = (avg_math2 + avg_reading2) /2
school_summary = pd.DataFrame({"School Type": [school_type], "Total Students": [total_students]
                                ,"Total School Budget": [total_budget2], "Per Student Budget": [per_student_budget],"Average Math Score": [avg_math2]
                                , "Average Reading Score": [avg_reading2]
                                ,"Percent of Sudents Passing Math": [percent_passing_math2]
                                , "Percent of Sudents Passing Reading": [percent_passing_reading2]
                                ,"Percent Overall Passing Rate": [overall_percent_passing2]})

school_summary[{"Total Students","Total School Budget","Per Student Budget","Average Math Score","Average Reading Score","Percent of Sudents Passing Math","Percent of Sudents Passing Reading","Percent Overall Passing Rate"}]
```

## Top Performing Schools (By % Overall Passing)

```python
# Top performing schools
school_summary = school_summary.sort_values(["Percent Overall Passing Rate"])

school_summary[["Total Students","Total School Budget","Per Student Budget","Average Math Score","Average Reading Score","Percent of Sudents Passing Math","Percent of Sudents Passing Reading","Percent Overall Passing Rate"]]
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Scores by School Spending
#bins
bins=[0,584,614,644,700]
row_name = ['>585','585 - 614', '615 - 644', '>644']
school_data_complete['spending'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], bins , labels = row_name)
gby_spending = school_data_complete.groupby('spending')
math_spend = gby_spending['math_score'].mean()
reading_spend = gby_spending['reading_score'].mean()
pass_math = school_data_complete[school_data_complete['math_score']>=70].groupby('spending')['Student ID'].count()/gby_spending['Student ID'].count()
pass_read = school_data_complete[school_data_complete['reading_score']>=70].groupby('spending')['Student ID'].count()/gby_spending['Student ID'].count()
overall = school_data_complete[(school_data_complete['reading_score']>=70) & (school_data_complete['math_score']>=70)].groupby('spending')['Student ID'].count()/gby_spending['Student ID'].count()
scorebyspend = pd.DataFrame({"Average Math Score": math_spend,
                            "Average Reading Score": reading_spend,
                            "% passing math": pass_math,
                            "% Passing Reading": pass_read,
                            "Overall Passing Rate": overall})
scorebyspend.index.name="Per Student Budget"
scorebyspend
```
